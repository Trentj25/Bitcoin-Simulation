# Bitcoin Simulation

```python
import pandas as pd
import random
```

Price and Volitility Data for Bitcoin, 5,000 days from one dollar to 100 thousand dollars

```python
input_csv = 'sim2.csv'
df = pd.read_csv(input_csv)
```

Simulating the impact on return of a Bitcoin portfolio when selling based on volitility
Data charges the rise of BT from $1 to $100K over 13 years from Nov 2011 to Dec 2024
Benchmark Group A, never sells
Group B (higher propensity to sell) only sell when conditions are met and rolls a pair of 1s with 4 sided dice
Group C only sells when volitility conditions are met and 1s are rolled on six sided dice

Group B --- 4 sided dice

```python
# 4 sided dice
random_min = 1
random_max = 17

#replicatability
random.seed(25)


# Output DataFrame 
output_data = {
    'First Value Index': [],
    'First Value': [],
    'Second Value': [],
    'Extra': [],
    'Iteration Count': []
}

# Is the difference between MV and BV > than volitility?  If so, throw dice. 

for i in range(len(df) - 1):  # Exclude last row
    loop_count = 0  # Initialize a counter for the inner loop
    broke = False  # Track if the loop broke
    for j in range(i + 1, len(df)):  # Compare with subsequent rows
        loop_count += 1  # Increment the loop counter
        difference = abs(df.loc[i, 'Value'] - df.loc[j, 'Value'])  # Absolute difference
        extra_value = df.loc[j, 'Extra']  # Fetch the corresponding pre-calculated volitility value at that date
        
        if difference > extra_value:  # Trigger random draw only if difference > volitlity
            draw = random.choice(range(random_min, random_max))  # Random draw
            if draw == 2:  # Break the loop only if roll a snake-eyes
                # Append results at break
                output_data['First Value Index'].append(df.loc[i, 'Index'])
                output_data['First Value'].append(df.loc[i, 'Value'])
                output_data['Second Value'].append(df.loc[j, 'Value'])
                output_data['Extra'].append(extra_value)
                output_data['Iteration Count'].append(loop_count) # number of days held
                broke = True
                break  # Exit and move to the next element in the outer loop

    if not broke:  # If the loop exhausts without breaking
        # Append the final comparison result
        output_data['First Value Index'].append(df.loc[i, 'Index'])
        output_data['First Value'].append(df.loc[i, 'Value'])
        output_data['Second Value'].append(df.loc[len(df) - 1, 'Value'])  # Last value in the column
        output_data['Extra'].append(df.loc[len(df) - 1, 'Extra'])  # Extra value corresponding to the last row
        output_data['Iteration Count'].append(loop_count)

# Create a DataFrame from the output data
output_df = pd.DataFrame(output_data)
```

```python
file_name = 'Output4Sim2.xlsx'
output_df.to_excel(file_name, index=False)
```
